# 二项分布  < Binomial >
## 日本語：二項分布（にこう ぶんぷ）

### 例题
80% of all the visitors to Lavista Museum end up buying souvenirs from the souvenir shop at the Museum. On the coming Sunday, if a random sample of 10 visitors is picked:

 1. Find the probability that every visitor will end up buying from the souvenir shop. 
 2. Find the probability that a maximum of 7 visitors will buy souvenirs from the souvenir shop.

```python
#パッケージをインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #可視化用
import seaborn as sns #可視化用
%matplotlib inline　
import scipy.stats as stats #科学計算用
```

```python
# vistorsの数量
n = 10
```

```python
# vistorsが購入する確率
p = 0.8
```

```python
# 10 visitorsに対して、様々な購入した人数の範囲（可能性:0-10人）
k = np.arange(0, 11)
```

```python
# 从SciPy里导入二项分布的函数（binom）
from scipy.stats import binom

# 使用二项分布函数（binom）来计算概略分布(因为是离散型变量，所以要用pmf)
binomial = binom.pmf(k=k, n=n, p=p)
```

```python
binomial #可以得出出10个人里0人，1人，2人..出现不同人数时对应的概率
```

```python
#図を描く
plt.bar(k, binomial)
plt.title("Binomial: n%i , p=%.2f" % (n,p), fontsize = 15)
plt.xlabel("Num of Success")
plt.ylabel("Probability of Success")
plt.show()
```

```python
binomial[10] #问题一：所有（10个）游客购买的概率
```

=======

#回答（根据上图）
Find the probability that every visitor will end up buying from the souvenir shop.
### 答案：0.107

=======

```python
#计算累计的概率
#本步骤只是为了解释“cdf”累计概率分布函数（概率的累计,红色的柱子），可以不加

barl = plt.bar(k, binomial)
plt.title("Binomial: n%i , p=%.2f" % (n,p), fontsize = 15)
plt.xlabel("Num of Success")
plt.ylabel("Probability of Success")

#选定7人以下
for i in range(0,8): 
    barl[i].set_color("r") #给这些柱状图上红色
plt.show()
```

```python
#利用cdf，计算累计分布函数
binom.cdf(k=7, n=n, p=p) #问题二：计算最多7人购买的概率（累计概率）
```

```python
#（拓展应用）利用cdf,计算4人以上购买的概率
1 - binom.cdf(k=3, n=n, p=p)
```

## 拓展：可以尝试其他的概率（P），来观察概率分布的变化

```python
plt.figure(figsize = (15,4)) #Figure size 1500x400

plt.subplot(131)
binomial_70 = binom.pmf(k, n, p=0.7)

#画出和上面相同的概率分布图
plt.bar(k, binomial_70)
plt.title("p=%.2f" % (0.7), fontsize = 15)
plt.xlabel("num of success")
plt.ylabel("Probability of success")

plt.subplot(132)
binomial_80 = binom.pmf(k, n, p=0.8)

#画出和上面相同的概率分布图
plt.bar(k, binomial_80)
plt.title("p=%.2f" % (0.8), fontsize = 15)
plt.xlabel("num of success")
plt.ylabel("Probability of success")

plt.subplot(133)
binomial_90 = binom.pmf(k, n, p=0.9)

#画出和上面相同的概率分布图
plt.bar(k, binomial_90)
plt.title("p=%.2f" % (0.9), fontsize = 15)
plt.xlabel("num of success")
plt.ylabel("Probability of success")

plt.tight_layout(w_pad = 5)
```
